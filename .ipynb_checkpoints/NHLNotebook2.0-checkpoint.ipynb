```python
# WANT: Proj SOG

# Model:

# - SOG/GP TY
# - SOG/GP L3-L5
# - Opp Team Allowed SOG/GP TY
# - Opp Team Allowed SOG/GP L3-L5

# Additional:

# - Any Starting Opp D out
# - Starting Opp G out
# - Back to back games
# - Home vs Away
# - Penalties

homeTeam = "EDM"
awayTeam = "BOS"
```

```python
# Resources:

# Imports
import pandas as pd
import numpy as np

# Endpoints
dataEndpoint = 'https://statsapi.web.nhl.com/api/v1/teams?expand=team.roster'

# Example: https://www.nhl.com/player/connor-mcdavid-8478402?stats=gamelogs-r-nhl&season=20222023
#
# Key: https://www.nhl.com/player/firstName-lastName-playerId?stats=gamelogs-r-nhl&season=20222023
resourceEndpoint = 'https://www.nhl.com/player/'
```

```python
import requests
import json

# Build Dataset

test = requests.get(dataEndpoint)

with open("data.json", "w") as outfile:
    json.dump(test.json(), outfile)

with open('data.json', 'r') as f:
    data = json.load(f)

data = data['teams']
```

```python
df = pd.DataFrame(columns=["team", "firstName", "lastName", "playerId"])

for team in data:
    abv = team['abbreviation']
    roster = team['roster']
    roster = roster['roster']
    for player in roster:
        person = player['person']
        fullName = person['fullName']
        fullName = fullName.split()
        firstName = fullName[0]
        lastName = fullName[1]
        id = person['id']
        
        entry = pd.DataFrame.from_dict({
            "team": [abv],
            "firstName":  [firstName],
            "lastName": [lastName],
            "playerId": [id]
        })
        
        df = pd.concat([df, entry], ignore_index=True)
        
df.loc[df['team'] == awayTeam]
```

```python
homedf = df.loc[df['team'] == homeTeam]
awaydf = df.loc[df['team'] == awayTeam]
gamedf = pd.concat([homedf, awaydf], ignore_index=True, sort=False)
gamedf
```

```python
fn = gamedf['firstName'][0]
ln = gamedf['lastName'][0]
id = gamedf['playerId'][0]

url = resourceEndpoint + fn + '-' + ln + '-' + str(id)

playerL5 = pd.read_html(url)
playerL5[5]
```
